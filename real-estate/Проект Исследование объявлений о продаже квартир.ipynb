## Предобработка данных

Пропуски есть в следующих столбцах:

Попытаемся их поочередно проанализировать и заполнить.

### Столбец `ceiling_height`
1. В столбце почти 40% значений пропущено. Скорее всего, многие продавцы просто не указывали этот параметр.
2. Возможно заполнить высоту потолков на основе **медианного значения**.
3. Примем во внимание, что, скорее всего, высота потолка должна находится в каких-то стандартных значениях. Например, стандартная высота потолков:
 * в "хрущевках" 2.5 м,
 * в "сталинках" 3-3.5 м,
 * в "брежневках" - 2.5-2.7,
 * в "панельках" - 2.65-2.75,
 * в современных домах 2.8-3.2 м.

Конечно, есть квартиры с высотой потолка более 4 м, но такую особенность указали бы при размещении объявления.

```python
for floors_total, median_ceiling_height in (data.query('2.5 < ceiling_height < 4')
                                            .groupby('floors_total')['ceiling_height']
                                            .median()
                                            .items()):
    data.loc[(data['ceiling_height'].isna())
             & (data['floors_total'] == floors_total),
             'ceiling_height'] = median_ceiling_height
```

```python
data['ceiling_height'].fillna(value=data.query('2.5 < ceiling_height < 3.2')['ceiling_height'].median(),
                              inplace=True)  # медианное значение возьмем для выборки от 2.5 до 3.2 м.
```

### Столбец `floors_total`
1. Пропущенных значений немного - скорее всего часть продавцов не указали данный параметр.
2. Заполнить пропуски можно **медианным значением**. Однако, необходимо рассчитать медиану в зависимости от этажа, на котором расположена квартира.
3. Необходимо заменить тип столбца на целочисленный, этажи всё же.

```python
for floor in data[data['floors_total'].isna()]['floor'].unique():  # перебираем этажи, для которых нет этажности дома
    data.loc[(data['floors_total'].isna())
             & (data['floor'] == floor),
             'floors_total'] = data[data['floor'] == floor]['floors_total'].median()
data['floors_total'] = data['floors_total'].astype('int')  # меняем тип данных столбца floors_total на целочисленный
```

Можно заметить, что все топонимы начинаются с большой буквы. Есть риск, что будут посёлок и деревня с одинаковыми названиями, но по беглому просмотру ничего, кроме Павлово, больше нет. Напишем функцию, которая оставляет только слова, начинающиеся с заглавной буквы и меняет ё на е.

```python
def istitle_only(string):
    title_word_list = []
    for word in string.split():
        if word.istitle():
            if 'ё' in word:
                word = word.replace('ё', 'е')
            title_word_list.append(word)
    return ' '.join(title_word_list)
```

```python
data.loc[:,['locality_name']] = data['locality_name'].apply(istitle_only)
```
